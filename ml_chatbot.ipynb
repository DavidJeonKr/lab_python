```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
sentences = ['안녕하세요?', '한국어 문장 임베딩을 위한 버트 모델입니다']
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/data/웰니스_대화_스크립트_데이터셋.xlsx')
```

```python
df.head()
```

```python
df = df[~df['챗봇'].isna()]
```

```python
df.head()
```

```python
df.loc[0, '유저']
```

```python
model.encode(df.loc[0, '유저'])
```

```python
df['embedding'] = pd.Series([[]] * len(df))

df['embedding'] = df['유저'].map(lambda x: list(model.encode(x)))
```

```python
df.head()
```

```python
df.to_csv('wellness_dataset.csv', index=False)
```

```python
text = '요즘 머리가 아프고 너무 힘들어'

embedding = model.encode(text)
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]

print('구분', answer['구분'])
print('유사한 질문', answer['유저'])
print('챗봇 답변', answer['챗봇'])
print('유사도', answer['distance'])
```
